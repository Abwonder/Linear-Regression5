### Exploratory Data Analysis

Dropping columns that will not be needed in this analysis, which are month, day and year

```python
### dropping this columns

df_copy = df.drop(["month", "day", "year"], axis = 1)
df_copy.head()
```

```python
## checking the categories for Classes

df_copy.Classes.value_counts()
```

```python
### use np.where from numpy to execute this!!!!!!!!

df_copy["Classes"] = np.where(df_copy["Classes"].str.contains("not fire"), 0,1)

# note: the structure of how it was created: not fire will be tagged 0 and 1 for fire
```

```python
## checking the categories for Classes

df_copy.Classes.value_counts()    ### sorted out now
```

```python
# checking the execution
df_copy.head()
```

```python
# checking the execution
df_copy.tail()
```

```python
## Plot desnity plot for all features
plt.style.use('seaborn')  ### This calls the use of seaborn here!!!
df_copy.hist(bins=50,figsize=(20,15))
plt.show()

##This show all the histogram for all the columns
```

```python
## Percentage for Pie Chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
percentage

## this is the way to calculate the percentage of each categories
```

```python
# plotting piechart
classlabels=["Fire","Not Fire"]  ### the arrangement of categories above
plt.figure(figsize=(12,9))  ## fig size
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
